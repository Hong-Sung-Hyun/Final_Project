```python
events= pd.read_csv('/Users/briansrebrenik/DSCourseMaterials/projects/finalProject/twitterData/cred_event_TurkRatings.data',
                   delimiter = '\t', error_bad_lines=False)
```

```python
events.Cred_Ratings = events.Cred_Ratings.apply(lambda x: [int(i) for i in re.findall('(-?\d)', x)])
```

```python
events['avg_rating'] = [np.average(x) for x in events.Cred_Ratings]
```

```python
events['median_rating'] = [np.median(x) for x in events.Cred_Ratings]
```

```python
events['more_than_half_less_than_2'] = [(x.count(1) + x.count(0) + x.count(-1) + x.count(-2)) >= 15 for x in events.Cred_Ratings]
```

```python
test = verified_tweets.groupby('user').count()
```

```python
test.drop(columns=['tweet_id', 'topic_id', 'date'], inplace=True)
test.rename(columns={'avg_score': 'tweet_count'}, inplace=True)
test.head()
```

```python
test2 = verified_tweets[verified_tweets.avg_score <= 1].groupby('user').count()
```

```python
test2.drop(columns=['tweet_id', 'topic_id', 'date'], inplace=True)
test2.rename(columns={'avg_score': 'tweet_count_below_1'}, inplace=True)
```

```python
scores = test.join(test2)
scores.head()
```

```python
scores.tweet_count_below_1.fillna(int(0), inplace=True)
```

```python
scores['percentage_of_tweets_below_1'] = scores.tweet_count_below_1 / scores.tweet_count
```

```python
more_than_25percent_below_1 = set(list(scores[scores.percentage_of_tweets_below_1 > .25]))
```

```python
scores['more_than_25percent_below_1'] = [x in more_than_25percent_below_1 for x in scores.index]
```

```python
scores.head()
```

```python
verified_users = pd.read_sql_table('verified_users', connection)
```

```python
verified_users.head()
```

```python
avg_tweet_score = verified_tweets.groupby('user').mean()
```

```python
avg_tweet_score = avg_tweet_score[['avg_score']]
```

```python
avg_tweet_score = avg_tweet_score.join(scores)
```

```python
avg_tweet_score['more_than_1_tweet_below_1'] = [x > 1.5 for x in avg_tweet_score.tweet_count_below_1]
```

```python
avg_tweet_score.drop(columns=['more_than_25percent_below_1', 'percentage_of_tweets_below_1'], inplace=True)
```

```python
avg_tweet_score['percentage_below_1'] = avg_tweet_score.tweet_count_below_1 / avg_tweet_score.tweet_count * 100
```

```python
avg_tweet_score['more_than_5_percent_fake'] = avg_tweet_score['percentage_below_1'] > 5
```

```python
percentile25= np.percentile(avg_tweet_score.avg_score, .25)
percentile50= np.percentile(avg_tweet_score.avg_score, .50)
print(percentile25, percentile50)
```

```python
avg_tweet_score['below_25th_percentile'] = avg_tweet_score.avg_score <= percentile25
```

```python
avg_tweet_score['below_50th_percentile'] = avg_tweet_score.avg_score <= percentile50
```

```python
node_table = verified_users.set_index('user').join(avg_tweet_score)
```

```python
node_table.head()
```

```python
node_table.reset_index(inplace=True)
node_table.set_index('user_id', inplace=True)
node_table.head()
```

```python
node_table.reset_index(inplace=True)
node_table.head()
```

```python
node_table.url.fillna(value='NA', inplace=True)
node_table.location.fillna(value='NA', inplace=True)
node_table.description.fillna(value='NA', inplace=True)
node_table.name.fillna(value='NA', inplace=True)
node_table.avg_score.fillna(value=0, inplace=True)
node_table.more_than_5_percent_fake.fillna(value='notweets', inplace=True)
node_table.below_25th_percentile.fillna(value='NA', inplace=True)
node_table.below_50th_percentile.fillna(value='NA', inplace=True)
node_table.below_25th_percentile = node_table.below_25th_percentile.astype(str)
node_table.below_50th_percentile = node_table.below_50th_percentile.astype(str)
node_table.more_than_5_percent_fake = node_table.more_than_5_percent_fake.astype(str)
#only include below code for creating gexf file
# for column in ['user', 'language', 'url', 'location', 'description', 'name', 'below_25th_percentile', 'below_50th_percentile', 'more_than_5_percent_fake']:
#     node_table[column] = node_table[column].astype(str)
#     node_table[column] = node_table[column].apply(lambda x: x.encode('ascii', errors='ignore'))
#     node_table[column] = node_table[column].astype(str)
# node_table.head()
```

```python
import csv
edge_list = []
with open('twitternetworkedges.csv', 'r') as f:
    data = csv.reader(f, delimiter=",", quotechar='"')
    next(data) # skips header row
    for row in data:
            edge_list.append(row)
```

```python
edge_list = [(int(x[0]), int(x[1])) for x in edge_list]
```

```python
G = nx.DiGraph()
```

```python
for i in tqdm(range(len(node_table))):
    G.add_node(int(node_table.node_id.iloc[i]), user=node_table.user.iloc[i], user_id=int(node_table.user_id.iloc[i]), friends=int(node_table.friends.iloc[i]),
              followers=int(node_table.followers.iloc[i]), language=node_table.language.iloc[i], url=node_table.url.iloc[i],
              location=node_table.location.iloc[i], description=node_table.description.iloc[i], 
              name=node_table.name.iloc[i], avg_score=float(node_table.avg_score.iloc[i]),
              below_25th_percentile=node_table.below_25th_percentile.iloc[i],
              below_50th_percentile=node_table.below_50th_percentile.iloc[i], more_than_5_percent_fake=node_table.more_than_5_percent_fake.iloc[i])
```

```python
G.add_edges_from(edge_list)
```

```python
nx.write_gexf(G, 'twitter_network.gexf')
```

```python
from neo4j import Driver
```

```python
driver = Driver(uri='bolt://localhost:7687', auth=('neo4j', 'password'))
```

```python
def add_nodes(driver):
    for i in tqdm(range(len(node_table))):
        test1 = node_table.node_id.iloc[i]
        test2 = node_table.user.iloc[i]
        test3 = node_table.friends.iloc[i]
        test4 = node_table.followers.iloc[i]
        test5 = node_table.avg_score.iloc[i]
        test6 = node_table.below_25th_percentile.iloc[i]
        test7 = node_table.below_50th_percentile.iloc[i]
        statement = "CREATE (:User { node_id: %s, user: '%s', friends: %s, followers: %s, avg_score: %s, below_25th_percentile: '%s', below_50th_percentile: '%s' })" % (test1, test2, test3, test4, test5, test6, test7)
        with driver.session() as session:
            session.run(statement)
```

```python
add_nodes(driver)
```

```python
with driver.session() as session:
    session.run('CREATE INDEX ON :User(node_id)')
```

```python
def add_edges(driver):
    for i in tqdm(range(len(edge_list))):
        node1 = edge_list[i][0]
        node2 = edge_list[i][1]
        statement = "MATCH (a:User {node_id: %s}), (b:User {node_id: %s}) CREATE (a)-[r:Following]->(b)" % (node1, node2)
        with driver.session() as session:
            session.run(statement)
```

```python
add_edges(driver)
```

```python
def add_below_1_attribute(driver):
    for i in tqdm(range(len(node_table))):
        node_id = node_table.node_id.iloc[i]
        below_1 = node_table.tweet_below_1.iloc[i]
        statement = "MATCH (a:User {node_id: %s}) SET a.tweet_below_1 = %s" % (node_id, below_1)
        with driver.session() as session:
            session.run(statement)
```

```python
add_below_1_attribute(driver)
```

```python
real_source = node_table[node_table.more_than_5_percent_fake == "False"].node_id
fake_source = node_table[node_table.more_than_5_percent_fake == 'True'].node_id
no_tweets = node_table[node_table.more_than_5_percent_fake == 'notweets'].node_id
```

```python
real_source = set(list(real_source))
fake_source = set(list(fake_source))
no_tweets = set(list(no_tweets))
print(len(real_source), len(fake_source), len(no_tweets))
```

```python
def fix_labels(driver):
    for i in tqdm(range(len(node_table))):
        node_id = node_table.node_id.iloc[i]
        if node_id in real_source:
            statement = "MATCH (a:User {node_id: %s}) SET a :RealSource" % (node_id)
            with driver.session() as session:
                session.run(statement)
        elif node_id in fake_source:
            statement = "MATCH (a:User {node_id: %s}) SET a :FakeSource" % (node_id)
            with driver.session() as session:
                session.run(statement)
        else:
            statement = "MATCH (a:User {node_id: %s}) SET a :NoTweets" % (node_id)
            with driver.session() as session:
                session.run(statement)
```

```python
fix_labels(driver)
```
